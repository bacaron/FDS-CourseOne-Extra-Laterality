```python
import pandas as pd
```

```python
df = pd.read_csv('./tractmeasures.csv')
df
```

```python
# brief summary statistics of the data
df.describe()
```

```python
# look at summary statistics only from camcan project
camcan = df.loc[df['classID'] == 'camcan']
camcan.describe()
```

```python
# look at per-tract average for everything in camcan
camcan.groupby('structureID').mean()
camcan.groupby('structureID').min()
camcan.groupby('structureID').max()
camcan.groupby('structureID').describe()
df.groupby('classID').mean()
```

```python
# look at per-tract, per classID average
df.groupby(['classID','structureID']).mean()
```

```python
# make plots
import seaborn as sns

###### make a scatterplot of tract FA between ping-siemens and camcan
# subset dataframe so we only have camcan and ping-siemens classes
ping_camcan = df.loc[(df['classID'] == 'ping-siemens') | (df['classID'] == 'camcan')]
ping_camcan

# groupby data by structureID and compute mean
ping_camcan_mean = ping_camcan.groupby(['classID','structureID']).mean().reset_index()

# generate scatterplot
sns.scatterplot(x=ping_camcan_mean.loc[ping_camcan_mean['classID'] == 'camcan']['fa'].tolist(),
                y=ping_camcan_mean.loc[ping_camcan_mean['classID'] == 'ping-siemens']['fa'].tolist(),
                hue=ping_camcan_mean.loc[ping_camcan_mean['classID'] == 'ping-siemens']['structureID'].tolist())
```

```python
# generate a scatterplot of FA vs MD
sns.scatterplot(x='fa',y='md',data=ping_camcan_mean,hue='structureID',legend=False)
```
